```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

df = pd.read_csv('/Users/mohmedpatel7/Downloads/movies.csv')
```

```python
df.head()
```

```python
# Looking for missing data
for col in df.columns:
    percent_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,percent_missing))
    
```

```python
df.dtypes
```

```python
# Changed NaN values to 0 and changed budget and gross income to int to remove decimal points.

df['budget'] = df['budget'].fillna(0)

df['gross'] = df['gross'].fillna(0)

df['budget']= df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
# Year and release date might have different years, so we create a new column to have the correct year.
df['correctyear'] = df['released'].str.extract('(\d{4})')

df.head()
```

```python
df = df.sort_values(by=['gross'], inplace = False, ascending = False)
```

```python
pd.set_option('display.max_rows',None)
```

```python
df.head()
```

```python
# Correlation Hypothesis - The higher the budget, the more likely they are to make more money (gross)
# We will test this by using a scatterplot

plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
df.head()
```

```python
sns.regplot(x='budget',y='gross',data = df,scatter_kws={"color":"red"},line_kws={"color":"blue"})
plt.show()
```

```python
df.corr(method='pearson',numeric_only= True)
```

```python
# There is a high correlation between budget and gross
# Let's make a heat map from our correlation matrix to visualize it better

correlation_matrix = df.corr(method='pearson',numeric_only= True)

sns.heatmap(correlation_matrix,annot = True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# Correlation Hypothesis 2 - There is a correlation between Company and Gross earnings.

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(method='pearson',numeric_only= True)

sns.heatmap(correlation_matrix,annot = True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
correlation_matrix = df_numerized.corr()

correlation_pairs = correlation_matrix.unstack()
correlation_pairs
```

```python
sorted_pairs = correlation_pairs.sort_values()

sorted_pairs
```

```python
high_correlation = sorted_pairs[(sorted_pairs) > 0.5]
high_correlation
```
